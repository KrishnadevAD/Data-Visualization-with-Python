#### Data visualization using matplotlib

##### Matploptib is a low-level library of Python which is used for data visualization. 
##### It is easy to use and emulates MATLAB like graphs and visualization. 
##### This library is built on the top of NumPy arrays and consist of several plots like line chart, 
##### bar chart, histogram, etc.

import numpy as np
import matplotlib.pyplot as plt
N=16
X=np.linspace(0,15,N)
Y=np.logspace(0.1,2,N)
print(Y)
plt.plot(X,Y,'o--')
plt.show()

```python
import numpy as np
import matplotlib.pyplot as plt
N=16
X=np.linspace(0,15,N)
Y=np.logspace(0.1,2000,N)
print(Y)
plt.plot(X,Y,'o--')
plt.show()
# series in the geometric progression
```

```python
#visualization of the quadratic graph
x=np.arange(30)
plt.plot(x,[(y**3 + 1)for y in x])
plt.show()
```

```python
#visualization of the quadratic graph can also be written as
x=np.arange(30)
plt.plot(x,x**3 + 1)
plt.show()
```

```python
#enable a grid in visualization
#manipulate the limit of axes
x=np.arange(3)
plt.plot(x,-x**2,x,-x**3,x,-2*x,x,-2**x)
plt.grid(True)
plt.savefig('test.png')#thsi saves the img in the current directory of the jupyter notebook
plt.show()
```

```python
# you can see the limit of axes by the axes  are set by default as follows
x=np.arange(3)
plt.plot(x,-x**2,x,-x**3,x,-2*x,x,-2**x)
plt.grid(True)
print(plt.axis())
plt.show()
```

```python
#using function xlim(), yim()
x=np.arange(3)
plt.plot(x,-x**2,x,-x**3,x,-2*x,x,-2**x)
plt.grid(True)
plt.xlim([0,2])
plt.ylim([-8,0])
print(plt.axis())
plt.show()
```

```python
#adding title and label for the axes

x=np.arange(3)
plt.plot(x,-x**2,x,-x**3,x,-2*x,x,-2**x)
plt.grid(True)
plt.xlabel('x=np.arange(3)------>')
plt.xlim([0,2])
plt.ylabel('y=f(x)------>')
plt.ylim([-8,0])
plt.title('simple plot Demooooo------>')
print(plt.axis())
plt.show()
```

```python
# you can pass an argument for the parameter label in the plot() function and then 
# call the function legend() to create a legend as follows:
import numpy as np 
import matplotlib.pyplot as plt
x=np.arange(3)
plt.plot(x,-x**2,label='-x**2')
plt.plot(x,-x**3,label='-x**3')
plt.plot(x,-2*x,label='-2*x')
plt.plot(x,-2**x,label='-2**x')
plt.legend()
plt.grid(True)
plt.xlabel('x=np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y=f(x)')
plt.ylim([-8,0])
plt.title("simple demo hai guys hehehe")
plt.show()
```

```python
 # instead of passing the legend string as an argument to the function plot()
# you can pass the ist of string as an argument to the legend()
x=np.arange(3)
plt.plot(x,-x**2,x,-x**3,x,-2*x,x,-2**x)
plt.legend(['-x**2','-x**3','-2*x','-2**x'])
plt.grid(True)
plt.xlabel('x=np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y=f(x)')
plt.ylim([-8,0])
plt.title("simple demo hai guys hehehe")
plt.show()
```

```python
# changing the location of the box
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(5)
plt.plot(x,-x**2,x,-x**3,x,-2*x,x,-2**x)
plt.legend(['-x**2','-x**3','-2*x','-2**x'],loc='lower center')# location is changed from here 
plt.grid(True)
plt.xlabel('x=np.arange(3)')
plt.xlim([0,2])
plt.ylabel('y=f(x)')
plt.ylim([-8,0])
plt.title("simple demo hai guys ")
plt.show()
```
